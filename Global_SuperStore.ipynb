```python
total_sales=df['Sales'].sum()
print(total_sales)
total_profit=df['Profit'].sum()
print(total_profit)
total_orders=df['Order_ID'].nunique()
print(total_orders)
total_customers=df['Customer_ID'].nunique()
print(total_customers)
profit_margin=(total_profit/total_sales)*100
print(profit_margin)
```

```python
df['Order Date']=pd.to_datetime(df['Order_Date'])   #converting object to datetime
```

```python
pd.options.display.float_format='{:,.2f}'.format
```

```python
#SALES TREND ANALYSIS
```

```python
df['year']=df['Order_Date'].dt.year     #created column YEAR
```

```python
df['month']=df['Order_Date'].dt.month_name
```

```python
df['year_month']=(df['Order_Date'].dt.to_period('M'))
```

```python
monthly_sales=df.groupby('year_month')['Sales'].sum().reset_index() 
monthly_sales.head()
monthly_sales['year_month']=monthly_sales['year_month'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(15, 6))
sns.lineplot(x='year_month', y='Sales', data=monthly_sales)
plt.title('Total Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

```python
df['quarter']=df['Order Date'].dt.quarter
```

```python
#how many transactions in each quarter 
df.groupby('quarter')['Order_ID'].count()
```

KEY FINDINGS - 

* Revenue grew gradually from 2011 - 2014
* 4th Quarter (OCT - DEC) generates the highest sale , possibly due to holiday season 
* TECHNOLOGY sector is the strongest category in Sales , Profit & Margin 
* Furniture category creates revenue but has weak profitability 
* Tables recorded a loss of around 65,000 causing the dip for furniture segment due to higher discounts .  
* A small number of customers contribute a significant share in revenue of the firm
* The most sales were made in the United States , while the least in Equatorial Guinea
* CONSUMER Segment is the company's highest revenue and profit generating segment .  
* Increased Discounts tend to lead to lower profits hence the pricing and discount distributions can be revised .
* Regional differences indicate opportunities for market expansion
* The business should prioritize profitable growth rather than revenue growth alone.
